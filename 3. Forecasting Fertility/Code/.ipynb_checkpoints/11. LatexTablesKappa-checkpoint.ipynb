```julia
function excel_to_latex_tables(excel_path, dir, years, captions)
    tables_dir = joinpath(dir, "tables")
    isdir(tables_dir) || mkdir(tables_dir)  # Ensure the directory exists

    workbook = XLSX.readxlsx(excel_path)
    sheet_names = filter(name -> occursin("Parameters", name), XLSX.sheetnames(workbook))

    # Ensure we do not exceed the bounds of the captions array
    num_sheets = min(length(sheet_names), length(captions))

    for index in 1:num_sheets
        sheet_name = sheet_names[index]
        sheet = workbook[sheet_name]
        caption = captions[index]
        label = "tab:parameters_k$index"

        # Initialize container for row data
        rows_data = []

        # Collect data for each specified row
        for row in 2:22
            kappa_mean = round(XLSX.getdata(sheet, row, 1), digits=3)
            kappa_lower = round(XLSX.getdata(sheet, row, 2), digits=3)
            kappa_upper = round(XLSX.getdata(sheet, row, 3), digits=3)
            push!(rows_data, [kappa_mean, kappa_lower, kappa_upper])
        end

        # Generate LaTeX code
        latex_code = generate_latex_table(rows_data, years, caption, label)
        file_path = joinpath(tables_dir, "table_kappa$index.tex")
        write(file_path, latex_code)
        println("LaTeX table for '$sheet_name' saved to $file_path")
    end
end

function generate_latex_table(rows_data, years, caption, label)
    latex_code = """
    \\begin{landscape}
    \\begin{table}[htbp]
    \\centering
    \\caption{$caption}
    \\label{$label}
    \\begin{tabular}{@{}lccc@{}}
    \\toprule
    Year & \\multicolumn{3}{c}{Parameter \\(\\kappa\\)} \\\\
    \\cmidrule(lr){2-4}
          & Mean Estimate & Lower 95\\% CI & Upper 95\\% CI \\\\
    \\midrule
    """
    
    for (i, year) in enumerate(years)
        row_data = rows_data[i]
        latex_code *= "$(year) & $(row_data[1]) & $(row_data[2]) & $(row_data[3]) \\\\\n"
    end

    latex_code *= "\\bottomrule\n\\end{tabular}\n\\end{table}\n\\end{landscape}"
    return latex_code
end


years = ["1998", "1999", "2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018"]
captions = ["Model A: Uneducated - Estimated Parameter kappa with Credible Intervals",
"Model A: Primary Education - Estimated Parameter kappa with Credible Intervals",
"Model A: Secondary Education - Estimated Parameter kappa  with Credible Intervals",
"Model A: Tertiary Education - Estimated Parameter kappa  with Credible Intervals",
"Model B: Uneducated - Estimated Parameter kappa  with Credible Intervals",
"Model B: Primary Education - Estimated Parameter kappa  with Credible Intervals",
"Model B: Secondary Education - Estimated Parameter kappa  with Credible Intervals",
"Model B: Tertiary Education - Estimated Parameter kappa  with Credible Intervals",
"Model C: Uneducated - Estimated Parameter kappa  with Credible Intervals",
"Model C: Primary Education - Estimated Parameter kappa  with Credible Intervals",
"Model C: Secondary Education - Estimated Parameter kappa  with Credible Intervals",
"Model C: Tertiary Education - Estimated Parameter kappa  with Credible Intervals"]

excel_to_latex_tables(excel_path, dir, years, captions)


```
