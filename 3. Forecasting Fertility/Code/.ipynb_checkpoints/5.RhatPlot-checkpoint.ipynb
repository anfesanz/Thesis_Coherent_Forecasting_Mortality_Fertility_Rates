```julia
using Turing, MCMCChains, Distributions, RData, CSV, DataFrames, LinearAlgebra, StatsFuns, Random, Serialization, Makie, CairoMakie, AlgebraOfGraphics
using AlgebraOfGraphics:density
using DataFrames:stack
dir = "/Users/felipesanchez/Documents/UoM/PhD/Phd Third Year/Chapter Fertility/ModellingF/Output/"

```

```julia
postdist = deserialize(dir * "chains_renamed_sum.jls")
#Modify String by Number (Float64) to plot
for (key, df) in postdist
    # Convert 'rhat' column from String to Float64 if not already done
    df.rhat = parse.(Float64, df.rhat)
end
```

```julia
function plot_rhat(postdist::Dict{Any, Any}, parameter_prefix::String, plot_dir::String)
    CairoMakie.activate!()  # Activate CairoMakie for static plots
    
    for (key, df) in postdist     
        # Filter rows for the given parameter prefix
        filtered_df = filter(row -> startswith(row.parameters, parameter_prefix), df)
        
        # Extract age ranges using a regex match from the 'parameters' column
        age_ranges = map(x -> match(r"\((.*?)\)", x).captures[1], filtered_df.parameters)
        
        # Plotting rhat values for the current DataFrame
        fig = Figure()
        ax = Axis(fig[1, 1], xlabel = "Age Range", ylabel = "Rhat")
        lines!(ax, 1:length(filtered_df.rhat), filtered_df.rhat, label = parameter_prefix)
        hlines!(ax, [1.1], color = :red, linestyle = :dash, label = "Threshold (1.1)")
        
        # Use the extracted age ranges for the x-axis tick labels
        ax.xticks = (1:length(age_ranges), age_ranges)
        
        # Set x-axis tick labels to be vertical for better readability
        ax.xticklabelrotation = 70
        
        # Save the plot to a file
        plot_path = joinpath(plot_dir, "Rhat_$(parameter_prefix)$(key).png")
        save(plot_path, fig)
    end
end
```

```julia
plot_rhat(postdist, "alpha", dir)
plot_rhat(postdist, "beta", dir)
plot_rhat(postdist, "kappa", dir)
```
