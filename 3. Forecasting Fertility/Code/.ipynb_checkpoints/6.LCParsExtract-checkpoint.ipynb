```julia
chain_model1f1 = deserialize(dir * "chain_model1f1.jls")
chain_model1f2 = deserialize(dir * "chain_model1f2.jls")
chain_model1f3 = deserialize(dir * "chain_model1f3.jls")
chain_model1f4 = deserialize(dir * "chain_model1f4.jls")
chain_model2f = deserialize(dir * "chain_model2f.jls")
chain_model3f = deserialize(dir * "chain_model3f.jls")

chains = [
    chain_model1f1,
    chain_model1f2,
    chain_model1f3,
    chain_model1f4,
    chain_model2f,
    chain_model3f] 

# Initialize an empty dictionary to hold the DataFrames
chains_sum = Dict()
for i in 1:length(chains)
    df = DataFrame(describe(chains[i])[1]) 
    chains_sum[i] = df
end
```

```julia
function extract_lcparametersA(chains_sum::Dict)
    lcparameters = Dict()
    
    # Regular expression pattern to match the desired parameters
    a_patterns = [r"a\[(\d+)\]", r"a1\[(\d+)\]", r"a2\[(\d+)\]", r"a3\[(\d+)\]", r"a4\[(\d+)\]"]

    for (key, df) in chains_sum
        # Initialize a dictionary to store vectors for each parameter pattern
        a_vectors = Dict(p => Float64[] for p in a_patterns)
        
        for row in eachrow(df)
            # Check each pattern and save the match
            for (pattern, vector) in a_vectors
                match_result = match(pattern, string(row.parameters))
                if !isnothing(match_result)
                    push!(vector, parse(Float64, string(row.mean)))
                end
            end
        end
        
        # Collect all non-empty vectors and save them under the key
        # Filtering out any empty vectors
        lcparameters[key] = [vector for vector in values(a_vectors) if !isempty(vector)]
    end
    
    return lcparameters
end

lcparametersA = extract_lcparametersA(chains_sum)

function extract_lcparametersB(chains_sum::Dict)
    lcparameters = Dict()
    
    # Regular expression pattern to match the desired parameters
    b_patterns = [r"b\[(\d+)\]"]

    for (key, df) in chains_sum
        # Initialize a dictionary to store vectors for each parameter pattern
        b_vectors = Dict(p => Float64[] for p in b_patterns)
        
        for row in eachrow(df)
            # Check each pattern and save the match
            for (pattern, vector) in b_vectors
                match_result = match(pattern, string(row.parameters))
                if !isnothing(match_result)
                    push!(vector, parse(Float64, string(row.mean)))
                end
            end
        end
        
        # Collect all non-empty vectors and save them under the key
        # Filtering out any empty vectors
        lcparameters[key] = [vector for vector in values(b_vectors) if !isempty(vector)]
    end
    
    return lcparameters
end

lcparametersB = extract_lcparametersB(chains_sum)

function extract_lcparametersK(chains_sum::Dict)
    lcparameters = Dict()
    
    # Regular expression pattern to match the desired parameters
    k_patterns = [r"k\[(\d+)\]", r"k1\[(\d+)\]", r"k2\[(\d+)\]", r"k3\[(\d+)\]", r"k4\[(\d+)\]"]

    for (key, df) in chains_sum
        # Initialize a dictionary to store vectors for each parameter pattern
        k_vectors = Dict(p => Float64[] for p in k_patterns)
        
        for row in eachrow(df)
            # Check each pattern and save the match
            for (pattern, vector) in k_vectors
                match_result = match(pattern, string(row.parameters))
                if !isnothing(match_result)
                    push!(vector, parse(Float64, string(row.mean)))
                end
            end
        end
        
        # Collect all non-empty vectors and save them under the key
        # Filtering out any empty vectors
        lcparameters[key] = [vector for vector in values(k_vectors) if !isempty(vector)]
    end
    
    return lcparameters
end

lcparametersK = extract_lcparametersK(chains_sum)

function extract_lcparametersL(chains_sum::Dict)
    lcparameters = Dict()

    # Regular expression patterns to match the desired parameters
    l_patterns = [
        r"logmu\[(\d+),(\d+)\]",
        r"logmu1\[(\d+),(\d+)\]",
        r"logmu2\[(\d+),(\d+)\]",
        r"logmu3\[(\d+),(\d+)\]",
        r"logmu4\[(\d+),(\d+)\]"
    ]

    # Assuming a fixed matrix size, or you could calculate max i, j from a preprocessing step
    max_i, max_j = 6, 21  # You need to adjust these based on your data

    for (key, df) in chains_sum
        # Create an OrderedDict of matrices with fixed sizes
        pattern_matrices = OrderedDict(p => zeros(Float64, max_i, max_j) for p in l_patterns)

        for row in eachrow(df)
            # Iterate over patterns in the defined order
            for (pattern, matrix) in pattern_matrices
                match_result = match(pattern, string(row.parameters))
                if !isnothing(match_result)
                    i, j = parse.(Int, match_result.captures)
                    if i <= max_i && j <= max_j  # Check to avoid out-of-bounds error
                        matrix[i, j] = parse(Float64, string(row.mean))
                    end
                end
            end
        end

        # Store non-empty matrices
        lcparameters[key] = [matrix for (p, matrix) in pattern_matrices if any(!iszero, matrix)]
    end

    return lcparameters
end

# Assuming `chains_sum` is defined elsewhere and passed to this function
lcparametersL = extract_lcparametersL(chains_sum)
```

```julia
logmu = Dict{Int, Matrix{Float64}}() 
logmu[1] = lcparametersL[1][1]
logmu[2] = lcparametersL[2][1]
logmu[3] = lcparametersL[3][1]
logmu[4] = lcparametersL[4][1]
logmu[5] = lcparametersL[5][1]
logmu[6] = lcparametersL[5][2]
logmu[7] = lcparametersL[5][3]
logmu[8] = lcparametersL[5][4]
logmu[9] = lcparametersL[6][1]
logmu[10] = lcparametersL[6][2]
logmu[11] = lcparametersL[6][3]
logmu[12] = lcparametersL[6][4]

kappa = Array{Any}(undef, 12)
kappa[1] = lcparametersK[1][1]
kappa[2] = lcparametersK[2][1]
kappa[3] = lcparametersK[3][1]
kappa[4] = lcparametersK[4][1]
kappa[5] = lcparametersK[5][1]
kappa[6] = lcparametersK[5][2]
kappa[7] = lcparametersK[5][3]
kappa[8] = lcparametersK[5][4]
kappa[9] = lcparametersK[6][1]
kappa[10] = lcparametersK[6][1]
kappa[11] = lcparametersK[6][1]
kappa[12] = lcparametersK[6][1]


beta = Array{Any}(undef, 12)
beta[1] = lcparametersB[1][1]
beta[2] = lcparametersB[2][1]
beta[3] = lcparametersB[3][1]
beta[4] = lcparametersB[4][1]
beta[5] = lcparametersB[5][1]
beta[6] = lcparametersB[5][1]
beta[7] = lcparametersB[5][1]
beta[8] = lcparametersB[5][1]
beta[9] = lcparametersB[6][1]
beta[10] = lcparametersB[6][1]
beta[11] = lcparametersB[6][1]
beta[12] = lcparametersB[6][1]


# Initialize the alpha array with sufficient space
alpha = Array{Any}(undef, 12)

# Assign the first four indices directly
alpha[1] = lcparametersA[1][1]
alpha[2] = lcparametersA[2][1]
alpha[3] = lcparametersA[3][1]
alpha[4] = lcparametersA[4][1]
alpha[5] = lcparametersA[5][1]
alpha[6] = lcparametersA[5][2]
alpha[7] = lcparametersA[5][3]
alpha[8] = lcparametersA[5][4]
alpha[9] = lcparametersA[6][1]
alpha[10] = lcparametersA[6][2]
alpha[11] = lcparametersA[6][3]
alpha[12] = lcparametersA[6][4]



lcparameters = Dict()

for i in 1:12
    if !haskey(lcparameters, i)
        lcparameters[i] = Dict()
    end
    lcparameters[i][1] = alpha[i]
    lcparameters[i][2] = beta[i]
    lcparameters[i][3] = kappa[i]
    lcparameters[i][4] = logmu[i]
end
```
