```julia
function mcmc_parameters(chain, param_a::String, param_b::String, param_k::String)
    # Extracting parameter `a`
    parameter = Symbol(param_a)
    a_samples = get(chain, parameter)[parameter]
    a_means = []
    a_credible_intervals = []
    for i in 1:length(a_samples)
        samples = a_samples[i]
        mean_value = mean(samples)
        push!(a_means, mean_value)
        lower, upper = quantile(samples, [0.025, 0.975]) # Calculating 95% credible interval
        push!(a_credible_intervals, (lower, upper))
    end

    # Extracting parameter `b`
    parameter = Symbol(param_b)
    b_samples = get(chain, parameter)[parameter]
    b_means = []
    b_credible_intervals = []
    for i in 1:length(b_samples)
        samples = b_samples[i]
        mean_value = mean(samples)
        push!(b_means, mean_value)
        lower, upper = quantile(samples, [0.025, 0.975]) # Calculating 95% credible interval
        push!(b_credible_intervals, (lower, upper))
    end

    # Extracting parameter `k`
    parameter = Symbol(param_k)
    k_samples = get(chain, parameter)[parameter]
    k_means = []
    k_credible_intervals = []
    for i in 1:length(k_samples)
        samples = k_samples[i]
        mean_value = mean(samples)
        push!(k_means, mean_value)
        lower, upper = quantile(samples, [0.025, 0.975]) # Calculating 95% credible interval
        push!(k_credible_intervals, (lower, upper))
    end

    return Dict("a_means" => a_means, "a_credible_intervals" => a_credible_intervals,
                "b_means" => b_means, "b_credible_intervals" => b_credible_intervals,
                "k_means" => k_means, "k_credible_intervals" => k_credible_intervals)
end
```

```julia
mcmc_parameters1 = mcmc_parameters(chain_model1f1, "a", "b", "k")
mcmc_parameters2 = mcmc_parameters(chain_model1f2, "a", "b", "k")
mcmc_parameters3 = mcmc_parameters(chain_model1f3, "a", "b", "k")
mcmc_parameters4 = mcmc_parameters(chain_model1f4, "a", "b", "k")
mcmc_parameters5 = mcmc_parameters(chain_model2f, "a", "b", "k")
mcmc_parameters6 = mcmc_parameters(chain_model2f, "a", "b", "k")
mcmc_parameters7 = mcmc_parameters(chain_model1m3, "a", "b", "k")
mcmc_parameters8 = mcmc_parameters(chain_model1m4, "a", "b", "k")
mcmc_parameters9 = mcmc_parameters(chain_model2f, "a1", "b", "k1")
mcmc_parameters10 = mcmc_parameters(chain_model2f, "a2", "b", "k2")
mcmc_parameters11 = mcmc_parameters(chain_model2f, "a3", "b", "k3")
mcmc_parameters12 = mcmc_parameters(chain_model2f, "a4", "b", "k4")
```

```julia
using XLSX

# Assuming mcmc_parameters1, mcmc_parameters2, ..., mcmc_parameters32 exist
# Create an array of these dictionaries for iteration
mcmc_parameters_array = [mcmc_parameters1, mcmc_parameters2, mcmc_parameters3, mcmc_parameters4, mcmc_parameters5, mcmc_parameters6, mcmc_parameters7, mcmc_parameters8, mcmc_parameters9, mcmc_parameters10, mcmc_parameters11, mcmc_parameters12, mcmc_parameters13, mcmc_parameters14, mcmc_parameters15, mcmc_parameters16, mcmc_parameters17, mcmc_parameters18, mcmc_parameters19, mcmc_parameters20, mcmc_parameters21, mcmc_parameters22, mcmc_parameters23, mcmc_parameters24, mcmc_parameters25, mcmc_parameters26, mcmc_parameters27, mcmc_parameters28, mcmc_parameters29, mcmc_parameters30, mcmc_parameters31, mcmc_parameters32]

function write_mcmc_ab_parameter_to_sheet!(sheet, row_offset, mcmc_parameter)
    for i in 1:length(mcmc_parameter["a_means"])
        # Column 1: a_means
        sheet[row_offset + i, 1] = mcmc_parameter["a_means"][i]
        
        # Columns 2 and 3: a_credible_intervals
        sheet[row_offset + i, 2] = mcmc_parameter["a_credible_intervals"][i][1]
        sheet[row_offset + i, 3] = mcmc_parameter["a_credible_intervals"][i][2]
        
        # Column 4: b_means
        sheet[row_offset + i, 4] = mcmc_parameter["b_means"][i]
        
        # Columns 5 and 6: b_credible_intervals
        sheet[row_offset + i, 5] = mcmc_parameter["b_credible_intervals"][i][1]
        sheet[row_offset + i, 6] = mcmc_parameter["b_credible_intervals"][i][2]
    end
end

function write_mcmc_k_parameter_to_sheet!(sheet, row_offset, mcmc_parameter)
    for i in 1:length(mcmc_parameter["k_means"])
        # Column 7: k_means
        sheet[row_offset + i, 1] = mcmc_parameter["k_means"][i]
        # Columns 8 and 9: k_credible_intervals
        sheet[row_offset + i, 2] = mcmc_parameter["k_credible_intervals"][i][1]
        sheet[row_offset + i, 3] = mcmc_parameter["k_credible_intervals"][i][2]
    end
end

function export_ab_excel(mcmc_parameters, dir)
    XLSX.openxlsx(dir * "mcmc_ab_parameters.xlsx", mode="w") do xf
        for (index, params) in enumerate(mcmc_parameters)
            sheet_name = "Parameters$(index)"
            sheet = XLSX.addsheet!(xf, sheet_name)
            
            write_mcmc_ab_parameter_to_sheet!(sheet, 1, params)
        end
    end
    println("Excel file 'mcmc_ab_parameters.xlsx' created successfully.")
end

function export_k_excel(mcmc_parameters, dir)
    XLSX.openxlsx(dir * "mcmc_k_parameters.xlsx", mode="w") do xf
        for (index, params) in enumerate(mcmc_parameters)
            sheet_name = "Parameters$(index)"
            sheet = XLSX.addsheet!(xf, sheet_name)
            
            write_mcmc_k_parameter_to_sheet!(sheet, 1, params)
        end
    end
    println("Excel file 'mcmc_k_parameters.xlsx' created successfully.")
end


export_ab_excel(mcmc_parameters_array, dir)
export_k_excel(mcmc_parameters_array, dir)
```
