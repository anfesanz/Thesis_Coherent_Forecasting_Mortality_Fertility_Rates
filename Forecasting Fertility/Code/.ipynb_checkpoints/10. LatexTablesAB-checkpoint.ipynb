```julia
function excel_to_latex_tables(excel_path, dir, age_groups, captions)
    tables_dir = joinpath(dir, "tables")
    isdir(tables_dir) || mkdir(tables_dir)  # Ensure the directory exists

    workbook = XLSX.readxlsx(excel_path)
    sheet_names = filter(name -> occursin("Parameters", name), XLSX.sheetnames(workbook))

    # Ensure we do not exceed the bounds of the captions array
    num_sheets = min(length(sheet_names), length(captions))

    for index in 1:num_sheets
        sheet_name = sheet_names[index]
        sheet = workbook[sheet_name]
        caption = captions[index]
        label = "tab:parameters_ab$index"

        # Initialize container for row data
        rows_data = []

        # Collect data for each specified row
        for row in 2:12
            alpha_mean = round(XLSX.getdata(sheet, row, 1), digits=3)
            alpha_lower = round(XLSX.getdata(sheet, row, 2), digits=3)
            alpha_upper = round(XLSX.getdata(sheet, row, 3), digits=3)
            beta_mean = round(XLSX.getdata(sheet, row, 4), digits=3)
            beta_lower = round(XLSX.getdata(sheet, row, 5), digits=3)
            beta_upper = round(XLSX.getdata(sheet, row, 6), digits=3)
            push!(rows_data, [alpha_mean, alpha_lower, alpha_upper, beta_mean, beta_lower, beta_upper])
        end

        # Generate LaTeX code
        latex_code = generate_latex_table(rows_data, age_groups, caption, label)
        file_path = joinpath(tables_dir, "table$index.tex")
        write(file_path, latex_code)
        println("LaTeX table for '$sheet_name' saved to $file_path")
    end
end

function generate_latex_table(rows_data, age_groups, caption,label)
    latex_code = "\\begin{landscape}\n\\begin{table}[htbp]\n\\centering\n\\caption{$caption}\n\\label{$label}\n\\begin{tabular}{@{}lcccccc@{}}\n\\toprule\nAge group & \\multicolumn{3}{c}{Parameter \\(\\alpha\\)} & \\multicolumn{3}{c}{Parameter \\(\\beta\\)} \\\\ \n\\cmidrule(lr){2-4} \\cmidrule(lr){5-7}\n          & Mean Estimate & Lower 95\\% CI & Upper 95\\% CI & Mean Estimate & Lower 95\\% CI & Upper 95\\% CI \\\\ \n\\midrule\n"

    for (i, age_group) in enumerate(age_groups)
        row_data = rows_data[i]
        latex_code *= "$(age_group) & $(row_data[1]) & $(row_data[2]) & $(row_data[3]) & $(row_data[4]) & $(row_data[5]) & $(row_data[6]) \\\\\n"
    end

    latex_code *= "\\bottomrule\n\\end{tabular}\n\\end{table}\n\\end{landscape}"
    return latex_code
end

age_groups = ["25-29", "30-34", "35-39", "40-44", "45-49", "50-54"]
captions = ["Model A: Uneducated - Estimated Parameters alpha and beta with Credible Intervals",
"Model A: Primary Education - Estimated Parameters alpha and beta with Credible Intervals",
"Model A: Secondary Education - Estimated parameters alpha and beta  with Credible Intervals",
"Model A: Tertiary Education - Estimated parameters alpha and beta  with Credible Intervals",
"Model B: Uneducated - Estimated parameters alpha and beta  with Credible Intervals",
"Model B: Primary Education - Estimated parameters alpha and beta  with Credible Intervals",
"Model B: Secondary Education - Estimated parameters alpha and beta  with Credible Intervals",
"Model B: Tertiary Education - Estimated parameters alpha and beta  with Credible Intervals",
"Model C: Uneducated - Estimated parameters alpha and beta  with Credible Intervals",
"Model C: Primary Education - Estimated parameters alpha and beta  with Credible Intervals",
"Model C: Secondary Education - Estimated parameters alpha and beta  with Credible Intervals",
"Model C: Tertiary Education - Estimated parameters alpha and beta  with Credible Intervals"]

excel_to_latex_tables(excel_path, dir, age_groups, captions)


```
