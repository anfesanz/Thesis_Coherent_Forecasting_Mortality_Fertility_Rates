```julia
using Turing, MCMCChains, Distributions, RData, CSV, XLSX, DataFrames, LinearAlgebra, StatsFuns, Random, Serialization, CairoMakie, AlgebraOfGraphics
using AlgebraOfGraphics:density
using DataFrames:stack
dir = "/Users/felipesanchez/Documents/UoM/PhD/Phd Third Year/Chapter Fertility/ModellingF/Output/"


function rename_par(chain)
    # Convert the chain to a DataFrame and save it
    df = DataFrame(chain)
    
    #Function to generate age labels
    age_label(i::Int) = i <= 5 ? "$(25 + 5*(i-1))-$(29 + 5*(i-1))" : "50-55"
    
    # Rename a[1] to a[6] and b[1] to b[6]
    for i in 1:6
        if hasproperty(df, Symbol("a[$i]"))
            rename!(df, Symbol("a[$i]") => Symbol("alpha($(age_label(i)))"))
        else
            for j in 1:4  # Loop over the 'a' groups from a1 to a4
                if hasproperty(df, Symbol("a$j[$i]"))
                    rename!(df, Symbol("a$j[$i]") => Symbol("alpha$j($(age_label(i)))"))
                end
            end
        end
        if hasproperty(df, Symbol("b[$i]"))
            rename!(df, Symbol("b[$i]") => Symbol("beta($(age_label(i)))"))
        end
    end

    # Rename k[1] to k[21] or handle k1[1] to k4[21] if k[i] does not exist
    for i in 1:21
        year = 1997 + i
        if hasproperty(df, Symbol("k[$i]"))
            rename!(df, Symbol("k[$i]") => Symbol("kappa($year)"))
        else
            for j in 1:4  # Loop over the 'k' groups from k1 to k4
                if hasproperty(df, Symbol("k$j[$i]"))
                    rename!(df, Symbol("k$j[$i]") => Symbol("kappa$j($year)"))
                end
            end
        end
    end
    # Rename logmu[x,y] from logmu[1,1] to logmu[11,21]
    for age in 1:6
        for year in 1:21
            old_name = Symbol("logmu[$age,$year]")
            new_name = Symbol("logmu[$(age_label(age)), $(1997 + year)]")
            if hasproperty(df, old_name) # Check if column exists before renaming
                rename!(df, old_name => new_name)
            end
        end
    end
    return df
end
        
function rename_par_sum(chain)
    # summary DataFrame
    df = DataFrame(describe(chain)[1])
    # Create a new DataFrame with all elements as strings
    dfs = DataFrame()
    for col in names(df)
        dfs[!, col] = string.(df[!, col])
    end
    for i in 1:size(dfs, 1)
        name = dfs[i, 1]  # Access the name in the first column
        
        # Match and replace a[i]
        if occursin(r"^a\[\d+\]$", name)
            index = parse(Int, match(r"\d+", name).match)
            age_range = index <= 10 ? "$(25 + 5*(index-1))-$(29 + 5*(index-1))" : "50-55"
            dfs[i, 1] = "alpha($age_range)"
        
        # Match and replace b[i]
        elseif occursin(r"^b\[\d+\]$", name)
            index = parse(Int, match(r"\d+", name).match)
            age_range = index <= 10 ? "$(25 + 5*(index-1))-$(29 + 5*(index-1))" : "50-55"
            dfs[i, 1] = "beta($age_range)"
        
        # Match and replace k[i]
        elseif occursin(r"^k\[\d+\]$", name)
            index = parse(Int, match(r"\d+", name).match)
            year = 1997 + index
            dfs[i, 1] = "kappa($year)"
        
        # Match and replace logmu[x,y]
        elseif occursin(r"^logmu\[\d+,\d+\]$", name)
            matches = match(r"logmu\[(\d+),(\d+)\]", name).captures
            age_index, year_index = parse(Int, matches[1]), parse(Int, matches[2])
            age_range = age_index <= 10 ? "$(25 + 5*(age_index-1))-$(29 + 5*(age_index-1))" : "50-55"
            year = 1997 + year_index - 1
            dfs[i, 1] = "logmu[$age_range, $year]"
        end
    end
    return dfs
end

function plot_chain(df::DataFrame, plot_path::String)
    CairoMakie.activate!() # Ensure CairoMakie is activated for saving plots
    
    # Check and create the plot path directory if it doesn't exist
    if !isdir(plot_path)
        mkpath(plot_path)
    end

    param_names = names(df)[3:end] # Adjust based on your DataFrame structure

    for param in param_names
        # Define the layers for the plot
        layers = data(df) * mapping(:iteration, Symbol(param)) * mapping(color = :chain => nonnumeric) * visual(Lines, alpha = 0.7)
        
        # Create the plot with AlgebraOfGraphics
        plt = draw(layers)
        
        # Save the plot to a file named after the parameter
        file_name = joinpath(plot_path, "$param.png")
        save(file_name, plt)  
    end
end
```

```julia
chain_model1f1 = deserialize(dir * "chain_model1f1.jls")
chain_model1f2 = deserialize(dir * "chain_model1f2.jls")
chain_model1f3 = deserialize(dir * "chain_model1f3.jls")
chain_model1f4 = deserialize(dir * "chain_model1f4.jls")
chain_model2f = deserialize(dir * "chain_model2f.jls")
chain_model3f = deserialize(dir * "chain_model3f.jls")

chains = [
    chain_model1f1,
    chain_model1f2,
    chain_model1f3,
    chain_model1f4,
    chain_model2f,
    chain_model3f]     
```

```julia
# Initialize an empty dictionary to hold the DataFrames
chains_renamed = Dict()
# Iterate through the chains array
id = 1
while id <= length(chains)
    # Paths
    chain_csv_path = dir * "chain_$id" * ".csv" 
    # Call your function to process the current chain and save the result
    df = rename_par(chains[id])
    
    # Store the DataFrame in the dictionary with the generated key
    chains_renamed[id] = df
    
    #Save CSV
    CSV.write(chain_csv_path, df)
    # Increment the id
    id += 1
end
```

```julia
# Initialize an empty dictionary to hold the DataFrames
chains_renamed_sum = Dict()
# Iterate through the chains array
id = 1
while id <= length(chains)
    # Paths
    chain_csv_sumpath = dir * "sum_chain_$id" * ".csv" 
    # Call your function to process the current chain and save the result
    dfs = rename_par_sum(chains[id])
    
    # Store the DataFrame in the dictionary with the generated key
    chains_renamed_sum[id] = dfs
    
    #Save CSV
    CSV.write(chain_csv_sumpath, dfs)
    # Increment the id
    id += 1
end
```

```julia
#saving the posterior means
serialize(dir * "chains_renamed_sum.jls", chains_renamed_sum)
```

```julia
#Doing the graphs
id = 1  # Initialize a counter for directory naming if id is not used from enumerate
while id <= length(chains_renamed)
    # Create a directory for the current chain's plots using id
    plot_path = mkpath(dir * "PlotChain$id/")   
    # Call function to process and save plots for the current chain DataFrame
    plot_chain(chains_renamed[id], plot_path)
    id += 1  # Increment counter after each iteration
end
```
