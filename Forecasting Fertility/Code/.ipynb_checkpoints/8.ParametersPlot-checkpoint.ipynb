```julia
using Plots
using Colors  

function plot_mcmc_parameters(chain, param_a::String, param_b::String, param_k::String; ylims_a=nothing, ylims_b=nothing, ylims_k=nothing)
    
    age_groups_labels = ["25-29", "30-34", "35-39", "40-44", "45-49", "50-54"]
    n_groups_age = length(age_groups_labels) 
    colors_age = cgrad(:cividis, n_groups_age, categorical=true)  # Color gradient for age groups

    # Extracting and plotting parameter `a`
    parameter = Symbol(param_a)
    a_samples = get(chain, parameter)[parameter]
    a_plot = Plots.plot(xlabel="Age Groups", ylabel=L"\alpha", xticks=(1:n_groups_age, age_groups_labels), xrotation=60, legend=false, grid=false, ylim=ylims_a)
    for (i, samples) in enumerate(a_samples)
        Plots.violin!(a_plot, [i for _ in 1:length(samples)], samples, label="", width=0.5, whisker_width=0.1, alpha=0.6, color=colors_age[i])
        mean_value = mean(samples)
        Plots.scatter!(a_plot, [i], [mean_value], color=:black, markersize=4, label="")
    end

    # Extracting and plotting parameter `b`
    parameter = Symbol(param_b)
    b_samples = get(chain, parameter)[parameter]
    b_plot = Plots.plot(xlabel="Age Groups", ylabel=L"\beta", xticks=(1:n_groups_age, age_groups_labels), xrotation=60, legend=false, grid=false, ylim=ylims_b)
    for (i, samples) in enumerate(b_samples)
        Plots.violin!(b_plot, [i for _ in 1:length(samples)], samples, label="", width=0.5, whisker_width=0.1, alpha=0.6, color=colors_age[i])
        mean_value = mean(samples)
        Plots.scatter!(b_plot, [i], [mean_value], color=:black, markersize=4, label="")
    end

    years_labels = ["1998", "1999", "2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018"]
    n_groups_years = length(years_labels)
    colors_years = cgrad(:cividis, n_groups_years, categorical=true)  # Color gradient for years

    # Extracting and plotting parameter `k`
    parameter = Symbol(param_k)
    k_samples = get(chain, parameter)[parameter]
    k_plot = Plots.plot(xlabel="Years", ylabel=L"\kappa", xticks=(1:n_groups_years, years_labels), xrotation=60, legend=false, grid=false, ylim=ylims_k)
    for (i, samples) in enumerate(k_samples)
        Plots.violin!(k_plot, [i for _ in 1:length(samples)], samples, label="", width=0.5, whisker_width=0.1, alpha=0.6, color=colors_years[i])
        mean_value = mean(samples)
        Plots.scatter!(k_plot, [i], [mean_value], color=:black, markersize=4, label="")
    end

    return a_plot, b_plot, k_plot
end

# To display or save the plots, you can use display(a_plot), display(b_plot), display(k_plot)
# or save them using savefig(a_plot, "a_plot.png") and so on
```

```julia
a1, b1, k1 = plot_mcmc_parameters(chain_model1f1, "a", "b", "k", ylims_a=(-10, 0), ylims_b=(0, 0.5), ylims_k=(-25, 20))
a2, b2, k2 = plot_mcmc_parameters(chain_model1f2, "a", "b", "k", ylims_a=(-10, 0), ylims_b=(0, 0.5), ylims_k=(-25, 20))
a3, b3, k3 = plot_mcmc_parameters(chain_model1f3, "a", "b", "k", ylims_a=(-10, 0), ylims_b=(0, 0.5), ylims_k=(-25, 20))
a4, b4, k4 = plot_mcmc_parameters(chain_model1f4, "a", "b", "k", ylims_a=(-10, 0), ylims_b=(0, 0.5), ylims_k=(-25, 20))
a5, b5, k5 = plot_mcmc_parameters(chain_model2f, "a1", "b", "k1", ylims_a=(-10, 0), ylims_b=(0, 0.5), ylims_k=(-25, 20))
a6, b6, k6 = plot_mcmc_parameters(chain_model2f, "a2", "b", "k2", ylims_a=(-10, 0), ylims_b=(0, 0.5), ylims_k=(-25, 20))
a7, b7, k7 = plot_mcmc_parameters(chain_model2f, "a3", "b", "k3", ylims_a=(-10, 0), ylims_b=(0, 0.5), ylims_k=(-25, 20))
a8, b8, k8 = plot_mcmc_parameters(chain_model2f, "a4", "b", "k4", ylims_a=(-10, 0), ylims_b=(0, 0.5), ylims_k=(-25, 20))
a9, b9, k9 = plot_mcmc_parameters(chain_model3f, "a1", "b", "k", ylims_a=(-10, 0), ylims_b=(0, 0.5), ylims_k=(-25, 20))
a10, b10, k10 = plot_mcmc_parameters(chain_model3f, "a2", "b", "k", ylims_a=(-10, 0), ylims_b=(0, 0.5), ylims_k=(-25, 20))
a11, b11, k11 = plot_mcmc_parameters(chain_model3f, "a3", "b", "k", ylims_a=(-10, 0), ylims_b=(0, 0.5), ylims_k=(-25, 20))
a12, b12, k12 = plot_mcmc_parameters(chain_model3f, "a4", "b", "k", ylims_a=(-10, 0), ylims_b=(0, 0.5), ylims_k=(-25, 20))
```
